```python
#siapa sih yg morning person ? 
pesan_pagi = df[(df['Jam_angka'] >= 4) & (df['Jam_angka'] <= 7)]
pengelompokan_chat_pagi = pesan_pagi.groupby('Tanggal').first()
hitung_chat_pagi = pengelompokan_chat_pagi['Pengirim'].value_counts()
print("siapa sih yg chatnya paling pagi ? ")
print(hitung_chat_pagi)
```

```python
chat_begadang = df[(df['Jam_angka'] >= 23) | (df['Jam_angka'] <= 3)]
penghitung_chat_begadang = chat_begadang['Pengirim'].value_counts()
print("siapa sih yg paling sering begadang diantara kita ? ")
print(penghitung_chat_begadang)
```

```python
total_per_jam = df['Jam_angka'].value_counts()
jam_sepi = total_per_jam.sort_values(ascending=True).head(5)
print("jam berapa sih chat paling sepi ? ")
print(jam_sepi)
```

```python

df['Waktu_Lengkap'] = pd.to_datetime(df['Tanggal'].astype(str) + ' ' + df['Jam'].astype(str), format='mixed', dayfirst=True)

df = df.sort_values('Waktu_Lengkap').reset_index(drop=True)

df['Selisih_Waktu'] = df['Waktu_Lengkap'].diff()


df['Apakah_Balasan'] = df['Pengirim'] != df['Pengirim'].shift(1)


data_balasan = df[df['Apakah_Balasan'] & df['Selisih_Waktu'].notna()]


waktu_balas = data_balasan.groupby('Pengirim')['Selisih_Waktu'].median()

print("siapa yg paling cepat balas chat?")
print(waktu_balas)
```

```python
slowrespon_chat= data_balasan.groupby('Pengirim')['Selisih_Waktu'].max()
print("siapa yg paling lama balas chat?")
print(slowrespon_chat)
```

```python
when_slowrespon = data_balasan.sort_values(by='Selisih_Waktu', ascending=False)


bukti_slowrespon= when_slowrespon[['Tanggal', 'Jam', 'Pengirim', 'Selisih_Waktu', 'Pesan']].head(5)

print("=== 5 MOMEN NGILANG TERLAMA (Bukti Sejarah) ğŸ“¸ ===")
print(bukti_slowrespon)
```

```python
# 1. Kita pinjam data utama kita (df) dan buat kolom untuk merekam "Pesan yang Dianggurin"
df['Pesan_Sebelumnya'] = df['Pesan'].shift(1)
df['Pengirim_Sebelumnya'] = df['Pengirim'].shift(1)

# 2. Kita filter ulang: Cari HANYA balasan AND jarak waktunya masuk akal (di bawah 1 hari)
data_konteks = df[df['Apakah_Balasan'] & (df['Selisih_Waktu'] < pd.Timedelta(days=1))]

# 3. Urutkan dari rekor ngilang terlama
rekor_lengkap = data_konteks.sort_values(by='Selisih_Waktu', ascending=False)

# 4. Susun tabelnya agar bercerita: Siapa yg ngetik apa -> Berapa lama ditinggal -> Dibalas apa
kolom_kronologi = ['Pengirim_Sebelumnya', 'Pesan_Sebelumnya', 'Selisih_Waktu', 'Pengirim', 'Pesan']
bukti_kronologi = rekor_lengkap[kolom_kronologi].head(5)

print("=== 5 KRONOLOGI NGILANG TERLAMA (Konteks Lengkap) ğŸ•µï¸â€â™‚ï¸ğŸ“¸ ===")
print(bukti_kronologi)
```
